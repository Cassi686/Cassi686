```python
import plotly.graph_objects as go

Close = go.Scatter(
                x = df.Date,
                y = df.Close,
                name = 'PETR4 Close',
                line = dict(color = '#17BECF'),
                opacity = 0.8)

data = [Close]

layout = dict(
    title = 'Faixa de Data Customizada',
    title_x= 0.5,
    xaxis = dict(
        range = ['2019-01-01', '2019-12-31'])
)

fig = dict(data = data, layout = layout)
py.iplot(fig, filename='Manually Set Range')
```

Rangerslider e Rangerselector

```python
Close = go.Scatter(
    x = df.Date,
    y = df.Close,
    name = 'PETR4 Hight',
    line = dict(color = '#17BECF'),
    opacity=0.8)

data = [Close]

layout = dict(
    title = 'Srie com Rangerslider e Botes ',
    title_x = 0.5,
    xaxis = dict(
        rangeselector = dict(
            buttons = list([
                dict(count=1,
                     label = '1m',
                     step = 'month',
                     stepmode = 'backward'),
                dict(count= 6,
                     label='6m',
                     step = 'month',
                     stepmode = 'backward'),
                dict(step='all')
            ])
        ),
        rangeslider = dict(
            visible = True
        ),
        type = 'date'
    )
)

fig = dict(data = data, layout = layout)
py.iplot(fig)
```

Trabalhando com Candlesticks

```python
trace = go.Candlestick(x=df['Date'],
                       open = df['Open'],
                       high = df['High'],
                       low = df['Low'],
                       close =df['Close'])
data = [trace]
py.iplot(data, filename='simple_candlestick')
```

Sem rangerslider

```python
trace = go.Candlestick(x=df['Date'],
                        open= df['Open'],
                        high=df['High'],
                        low=df['Low'],
                        close=df['Close'])
layout = go.Layout(
    title='Preos de Candlesticks',
    title_x=0.5,
    xaxis= dict(
        rangeslider=dict(
            visible=False
        )
    )
)

data=[trace]

figure=go.Figure(data=data,layout=layout)
py.iplot(fig)
```

```python
df['MM9']= df.Close.rolling(window=9).mean() # Mdia de fechamento dos ltimos 9 dias

df['MM17']= df.Close.rolling(window=17).mean()#Mdia de fechamento dos ltios 17 dias
```

```python
df.head(30)
```

Plot da mdia de preos

```python
close = go.Scatter(
                x= df.Date,
                y=df.Close,
                name='PETR4 Close',
                line=dict(color='#330000'),
                opacity=0.8)

MM_9 = go.Scatter(
               x=df.Date,
               y=df['MM9'],
               name='Mdia mvel 9 perodos',
               line=dict(color='#B2FF66'),
               opacity=0.8)

MM_17 = go.Scatter(
                x=df.Date,
                y=df['MM17'],
                name='Media mvel 17 perodos',
                line=dict(color='#FF00FF'),
                opacity=0.8)

data = [close,MM_9,MM_17]

py.iplot(data)
```
