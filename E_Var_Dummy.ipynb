```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('/content/C├│pia de Limite_Credito_Dummy.xlsx')
df.head()
```

```python
df.rename(columns={'Gênero':'Região', 'Região':'Gênero'},inplace= True)
```

```python
df.head()
```

```python
df = pd.get_dummies(data = df, columns = ["Escolaridade", "Gênero", "Região"] , prefix=["Escolaridade", "Gênero", "Região"])
df.head()
```

```python
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Gênero_F+Gênero_M+Região_RJ+Região_SP'
model = smf.ols(formula=function, data=df).fit()
print(model.summary())
```
